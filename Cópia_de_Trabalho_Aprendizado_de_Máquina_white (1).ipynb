Remo√ß√£o dos atributos 'density' e 'total sulfur dioxide'

```python
df = df.drop(columns=['density', 'total sulfur dioxide'])
```

normaliza√ß√£o + SMOTE + modelo -> com 5-fold cross-validation estratificado



```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE

# separa features e target
X = df.drop(columns=['quality', 'quality_binary'])
y = df['quality_binary']

modelos = {
    "Regress√£o Log√≠stica": LogisticRegression(max_iter=200),
    "√Årvore de Decis√£o": DecisionTreeClassifier(),
    "MLP": MLPClassifier(max_iter=500),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier()
}

# k-fold estratificado
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=37)

# fun√ß√£o p/ rodar spot-checking
def avaliar_modelos(modelos):
    resultados = {}

    for nome, modelo in modelos.items():

        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('smote', SMOTE(random_state=37)),
            ('model', modelo)
        ])

        accuracy_scores  = cross_val_score(pipeline, X, y, cv=skf, scoring='accuracy')
        f1_scores        = cross_val_score(pipeline, X, y, cv=skf, scoring='f1')
        precision_scores = cross_val_score(pipeline, X, y, cv=skf, scoring='precision')
        recall_scores    = cross_val_score(pipeline, X, y, cv=skf, scoring='recall')

        resultados[nome] = {
            "Accuracy":  (accuracy_scores.mean(), accuracy_scores.std()),
            "F1":        (f1_scores.mean(), f1_scores.std()),
            "Precision": (precision_scores.mean(), precision_scores.std()),
            "Recall":    (recall_scores.mean(), recall_scores.std()),

            # valores brutos para gr√°ficos ‚Äî ESSENCIAL
            "Accuracy_raw": accuracy_scores,
            "F1_raw": f1_scores,
            "Precision_raw": precision_scores,
            "Recall_raw": recall_scores
        }

        print(f"\n=== {nome} ===")
        print(f"Accuracy m√©dio:  {accuracy_scores.mean():.4f} (¬± {accuracy_scores.std():.4f})")
        print(f"F1 m√©dio:        {f1_scores.mean():.4f} (¬± {f1_scores.std():.4f})")
        print(f"Precis√£o m√©dia:  {precision_scores.mean():.4f} (¬± {precision_scores.std():.4f})")
        print(f"Recall m√©dio:    {recall_scores.mean():.4f} (¬± {recall_scores.std():.4f})")

    return resultados

resultados = avaliar_modelos(modelos)
```

```python
!pip install joypy
import seaborn as sns
import joypy

# lista para armazenar registros individuais
linhas = []

for modelo, metricas in resultados.items():
    for i in range(5):  # 5 folds da valida√ß√£o cruzada
        linhas.append({
            "Modelo": modelo,
            "Accuracy":  metricas["Accuracy_raw"][i]  if "Accuracy_raw"  in metricas else None,
            "F1":        metricas["F1_raw"][i]        if "F1_raw"        in metricas else None,
            "Precision": metricas["Precision_raw"][i] if "Precision_raw" in metricas else None,
            "Recall":    metricas["Recall_raw"][i]    if "Recall_raw"    in metricas else None
        })

df_scores = pd.DataFrame(linhas)
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_scores, x="Modelo", y="Accuracy")
plt.title("Distribui√ß√£o da Accuracy por Modelo")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14, 8))
sns.boxplot(data=df_scores, x="Modelo", y="F1")
plt.title("Distribui√ß√£o do F1-score por Modelo")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14, 8))
sns.boxplot(data=df_scores, x="Modelo", y="Precision")
plt.title("Distribui√ß√£o da Precis√£o por Modelo")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14, 8))
sns.boxplot(data=df_scores, x="Modelo", y="Recall")
plt.title("Distribui√ß√£o do Recall por Modelo")
plt.xticks(rotation=45)
plt.show()
```
